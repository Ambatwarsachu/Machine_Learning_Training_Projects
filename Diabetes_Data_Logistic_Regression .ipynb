```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/training-ml/Files/main/diabetes.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
#Data Cleaning
#replacing zero values with the mean of the column
data['BMI'] = data['BMI'].replace(0, data['BMI'].mean())
data['Glucose'] = data['Glucose'].replace(0, data['Glucose'].mean())
data['BloodPressure'] = data['BloodPressure'].replace(0, data['BloodPressure'].mean())
data['SkinThickness'] = data['SkinThickness'].replace(0, data['SkinThickness'].mean())
data['Insulin'] = data['Insulin'].replace(0, data['Insulin'].mean())
```

```python
data.describe()
```

```python
#lets see how data is distributed for every column
plt.figure(figsize=(20,25),facecolor='green')
plotnumber = 1
for column in data:
    if plotnumber <=9:
        ax=plt.subplot(3,3,plotnumber)
        pl=sns.distplot(data[column])
        plt.xlabel(column, fontsize=20)
        plt.ylabel(column, fontsize=20)
    plotnumber+=1
plt.show()
```

```python
bp_high = (q3.BloodPressure + (1.5*iqr.BloodPressure))
print (bp_high)

bp_index = np.where(data['BloodPressure'] > bp_high)

data = data.drop(data.index[bp_index])
print(data.shape)

data.reset_index() # now the 10 index are outliers and we are select and deleated that outliers
```

```python
gl_high = (q3.Glucose + (1.5*iqr.Glucose))
print(gl_high)

gl_index = np.where(data['Glucose']>gl_high)

data = data.drop(data.index[gl_index])
print(data.shape)
data.reset_index()
```

```python
sk_high = (q3.SkinThickness + (1.5*iqr.SkinThickness))
print(sk_high)

sk_index = np.where(data['SkinThickness'] > sk_high)

data = data.drop(data.index[sk_index])
print(data.shape)
data.reset_index()
```

```python
ins_high = (q3.Insulin+(1.5*iqr.Insulin))
print(ins_high)

ins_index = np.where(data['Insulin']>ins_high)

data= data.drop(data.index[ins_index])
print(data.shape)
data.reset_index()
```

```python
bmi_high = (q3.BMI+(1.5*iqr.BMI))
print(bmi_high)

bmi_index = np.where(data['BMI']>bmi_high)

data = data.drop(data.index[bmi_index])
print(data.shape)
data.reset_index()
```

```python
dpf_high = (q3.DiabetesPedigreeFunction + (1.5*iqr.DiabetesPedigreeFunction))
print(dpf_high)

dpf_index = np.where(data['DiabetesPedigreeFunction']>dpf_high)

data = data.drop(data.index[dpf_index])
print(data.shape)
data.reset_index()
```

```python
age_high = (q3.Age + (1.5*iqr.Age))
print(age_high)

age_index = np.where(data['Age']>age_high)

data = data.drop(data.index[age_index])
print(data.shape)
data.reset_index()
```

```python
bp_low = (q1.BloodPressure-(1.5*iqr.BloodPressure))
print(bp_low)

bp_low_index = np.where(data['BloodPressure']<bp_low)

data = data.drop(data.index[bp_low_index])
print(data.shape)
data.reset_index()
```

```python
plt.figure(figsize=(20,25),facecolor='yellow')
plotnumber = 1
for column in data:
    if plotnumber <=9:
        ax=plt.subplot(3,3,plotnumber)
        pl=sns.distplot(data[column])
        plt.xlabel(column, fontsize=20)
    plotnumber+=1
plt.show()
```

```python
x_x = data.drop(columns=['Outcome'])
y_y = data['Outcome']
```

```python
#before fit our data into a model, lets visualize the relationshop between our indeoendent variables and the categerical 
#lets see how features are related to class
plt.figure(figsize = (15,20))
plotnumber = 1
for column in x_x:
    if plotnumber <=9:
        ax = plt.subplot (3,3,plotnumber)
        sns.stripplot(x=y_y, y=x_x[column], hue=y_y)
    plotnumber+=1
plt.show()
```

```python
#it indicated that as glucose level increases there is s high cjance of that ou are giving to be a diabtic paitent
#If bloodpressure increases the chance of being a diabetic is increasing
#Cheking Multicollinearity problem
#one feature is dependent on another feature , feature vs feature
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_x)
```

```python
x_scaled.shape[1]
```

```python
#finfing variacne inflation factor in each scaled column 
vif = pd.DataFrame()
vif['vif']= [variance_inflation_factor(x_scaled, i ) for i in range(x_scaled.shape[1])]
vif['features']= x_x.columns
vif
```

```python
#one by one it is going to compare with other column and it is come up with the score. uses formula 1/1-r2
#what above score indicates--> it depends on project to project
#if the score is vif is less than 5 then it is considered as there is no muticollinearity(client will give that number)
# lets split our data in test and traing set
x_train,x_test,y_train,y_test = train_test_split(x_scaled, y_y, test_size=0.25, random_state=355)
```

```python
#Pre-processing stes are complted
#Model Building starts
log_reg= LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
y_pred = log_reg.predict(x_test)
y_pred
```

```python
log_reg.predict_proba(x_test)
```

```python
#confusion matrix
conf_mat = confusion_matrix(y_test,y_pred)
conf_mat
```

```python
#model accuracy
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#roc curve
fpr,tpr, thresholds = roc_curve(y_test,y_pred)
#thresholds[0]means no instance is predicted(it should be read from 0-max)
print('Theshold=', thresholds)
print('True Positive Rate=', tpr)
print('False positive Rate=', fpr)
```

```python
plt.plot(fpr, tpr, color='orange', label='ROC')
plt.plot([0,1],[0,1],color='darkblue', linestyle='--')
plt.xlabel('fasle positive rate')
plt.ylabel('true positive rate')
plt.title('receiver operating characteristic (ROC) curev')
plt.legend()
plt.show
```

```python
auc_score=roc_auc_score(y_test,y_pred)
print(auc_score)
```
